```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
miles_df = pd.read_csv('datos/us-airlines-monthly-aircraft-miles-flown.csv',header = 0, parse_dates=[0])
```

```python
miles_df.head()
```

```python
miles_df.index = miles_df['Month']
```

```python
miles_df.head()
```

```python
result = seasonal_decompose(miles_df['MilesMM'],model='additive')
```

```python
result.plot()
```

```python
miles_df.plot()
```

```python
result2 = seasonal_decompose(miles_df['MilesMM'],model='multiplicative')
```

```python
result2.plot()
```
